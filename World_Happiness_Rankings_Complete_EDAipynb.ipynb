#Happiest and Unhappiest Countries in 2024

```python
df2024_happiest_unhappiest = df2024[(df2024.loc[:,'Ladder score']>7.4)|(df2024.loc[:,'Ladder score']<3.5)]
sns.barplot(x = 'Ladder score', y= 'Country name',data = df2024_happiest_unhappiest, palette = 'coolwarm')
plt.title('Happiest and Unhappiest Countries in 2024')
plt.show()
```

######In this section, we created a chart displaying the happiest and unhappiest countries by using the Ladder Score value. While determining the boundaries here, we took into account the chart we created in the previous step. We set two limits according to the lower and upper limits in the graph of the ladder score.
If we need to make an inference from here; We can say that Western European countries are happier than other countries. In the following section, we will re-examine whether this conclusion is correct or not by making a region-based ranking.

#Ladder Score Distribution by Regional Indicator

```python
plt.figure(figsize = (15,8))
sns.kdeplot(x=df2024['Ladder score'], hue = df2024['Regional indicator'], fill = True, linewidth = 2)
plt.axvline(df2024['Ladder score'].mean(),c= 'black')
plt.title('Ladder Score Distribution by Regional Indicator')
plt.show()
```

####In this section, we compared the ladder score value on a region-by-region basis by using a different visualization format.

According to the chart; We can say that happiness is low in the African region,South Asia and the happiness rate is high in Western Europe. This inference also supports the inference we made from the previous graph.

#Relationship Between Features

```python
non_numeric_columns = df.select_dtypes(exclude=['float64', 'int64']).columns
df_numeric = df.drop(columns=non_numeric_columns)
correlation_matrix = df_numeric.corr()
sns.heatmap(correlation_matrix, annot = True, fmt ='.2f', linewidth = .7)
plt.title('Relationship Between Features')
plt.show()
```

####If we look at where the highest relationship is in inverse proportion; We see that there is a relationship between corruption and freedom with a value of 0.46. We also found this inverse proportionality relationship by interpreting it on the map in the previous step.

Let's look at how Life Ladder is proportional to which features. We see that the highest ratio is with national income, with a value of 0.77. We can then comment that it is linked to social support and then to health.

```python
sns.clustermap(correlation_matrix, center = 0, cmap = 'vlag', dendrogram_ratio = (0.1,0.2), annot = True, linewidth = .7, figsize=(10,10))
plt.show()
```

###Parameters Explained:
sns.clustermap(...): This function from the Seaborn library creates a clustermap, which is a type of heatmap that also performs hierarchical clustering on both rows and columns. It shows the data matrix with a dendrogram (tree-like diagram) that reflects the clustering.

correlation_matrix: This is the data that you want to visualize. In this case, it is presumably a matrix of correlation values, which can be visualized in a heatmap format with hierarchical clustering applied.

##Parameters:
center = 0: Centers the color map around zero.

cmap = 'vlag': Uses the 'vlag' colormap for the heatmap.

dendrogram_ratio = (0.1, 0.2): Controls the size of the dendrograms.

annot = True: Displays cell values on the heatmap.

linewidth = .7: Sets the thickness of lines between cells.

figsize = (10, 10): Sets the size of the figure.
